```python
# Implementation of image Smoothing using mean Filter

import cv2
import matplotlib.pyplot as plt
# Step 1: Load image (grayscale for simplicity)
img = cv2.imread("imp.jpeg", cv2.IMREAD_GRAYSCALE)
# Step 2: Apply Mean Filter (Averaging)
# Parameters: (image, kernel_size)
mean_blur = cv2.blur(img, (5,5)) # 5x5 mean filter, kernel size 5 x 5
# Step 3: Display Results
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img, cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(mean_blur, cmap="gray")
plt.title("Mean Filtered Image")
plt.axis("off")
plt.show()
```

```python
# Implementation of image Smoothing using Gaussian Filter

import cv2
import matplotlib.pyplot as plt
# Step 1: Read Image (grayscale for simplicity)
img = cv2.imread("imp.jpeg", cv2.IMREAD_GRAYSCALE)
# Step 2: Apply Gaussian Filter
# Parameters: (image, kernel_size, sigma)
gaussian_blur = cv2.GaussianBlur(img, (5,5), 1) # σ=1, Small σ → less blur, keeps detail. Large σ → moreblur, strong smoothing.
# Step 3: Display Results
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img, cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(gaussian_blur, cmap="gray")
plt.title("Gaussian Smoothed Image")
plt.axis("off")
plt.show()
```

```python
# Implementation of image sharpening using Laplacian Filter

import cv2
import numpy as np
import matplotlib.pyplot as plt
# Step 1: Load image (grayscale for simplicity)
img = cv2.imread("imp.jpeg", cv2.IMREAD_GRAYSCALE)
# Step 2: Apply Laplacian Filter
laplacian = cv2.Laplacian(img, cv2.CV_64F) # Compute Laplacian
laplacian = cv2.convertScaleAbs(laplacian) # Convert back to uint8
# Step 3: Sharpen Image (Original + Laplacian)
sharpened = cv2.addWeighted(img, 1.0, laplacian, 1.0, 0)
# Step 4: Display Results
plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.imshow(img, cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(laplacian, cmap="gray")
plt.title("Laplacian (Edges)")
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(sharpened, cmap="gray")
plt.title("Sharpened Image")
plt.axis("off")
plt.show()
```
